```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = range(100)
Y = [value ** 2 for value in X]
plt.plot(X, Y)
plt.show()
```

```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]
X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 5)
ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)
ax_1.plot(X_1, Y_1, c="b")
ax_2.plot(X_2, Y_2, c="g")
plt.show()
```

```python
plt.title("$y = \\frac{ax+b}{test}$")
plt.plot(range(100), range(100), label="line_1")
plt.legend(shadow=True, fancybox=True, loc="lower right")
plt.show()
```

```python
import numpy as np
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x")
plt.scatter(data_2[:,0], data_1[:,1], c="r", marker="^")

plt.show()
```

```python
N = 50
p = np.random.rand(N, 2)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2

plt.scatter(p[:,0], p[:,1], s=area, c=colors, alpha=0.5)
plt.show()
```
